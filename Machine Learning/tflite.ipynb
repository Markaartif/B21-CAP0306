#Convert Model to TFLite

```python
import tensorflow as tf

# Load Saved Model and Initialize the Converter
converter = tf.lite.TFLiteConverter.from_saved_model('drive/MyDrive/Colab Notebooks/BANGKIT/mobilenet')

# Set Optimzations
converter.optimizations = [tf.lite.Optimize.OPTIMIZE_FOR_LATENCY]

# Generate TFLite Model
tflite_model = converter.convert()
```

```python
import pathlib

#Save TFLite Model
tflite_model_file = pathlib.Path('drive/MyDrive/Colab Notebooks/BANGKIT/model.tflite')
tflite_model_file.write_bytes(tflite_model)
```
